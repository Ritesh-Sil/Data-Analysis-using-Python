```python
df_raw = pd.read_csv('C:\\Users\\silri\\Applied_AI\\EDA\DataSets\\full_data.csv')
```

```python
df = df_raw
```

```python
df['avg_bmi_by_age'] = df[['age','bmi']].groupby(['age']).mean()
```

##### Find the distribution of age

```python
fig = plt.figure(figsize=(8,6))
sns.set_style('darkgrid')
sns.kdeplot(data = df, x = 'age', fill = True, alpha = 0.1, label = 'Age' )
sns.kdeplot(data = df, x = 'avg_glucose_level', fill = True, alpha =0.15 , label = 'Avg Glucose level')
plt.legend()
```

```python
#### We want to see avg_glucose_level and Bmi by age
```

```python
fig = plt.figure(figsize=(16,10))
fig.tight_layout(w_pad = 8)
ax1 = fig.add_subplot(2,2,1)
sns.scatterplot(data=df, x = 'age', y = 'bmi', color = '#c36b98', alpha = 0.3)
sns.lineplot(data=df, x = 'age', y = 'avg_bmi_by_age', color = 'black', alpha = 0.6,marker = 'o')
ax1.set_xlabel('age')
ax1.set_ylabel('bmi')

ax2 = fig.add_subplot(2,2,2)
sns.scatterplot(data = df, x = 'age', y = 'avg_glucose_level', color = '#575f2a', alpha = 0.3)
ax2.set_xlabel('age')
ax2.set_ylabel('avg_glucose_level')
```

```python
## Plot the age and glucose level distribution in separate plots
```

```python
fig = plt.figure(figsize=(20,15))
ax1 = fig.add_subplot(2,2,1)
sns.kdeplot(data = df, x ='age', color = '#643296', fill =True , alpha = 0.15, hue='gender')
ax1.set_title('Age distribution')
ax1.set_xlabel('Age', fontsize =14)
ax1.set_ylabel('Density', fontsize = 14)


ax2 = fig.add_subplot(2,2,2)
sns.kdeplot(data = df, x ='bmi', color = '#643296', fill =True , alpha = 0.15, label = 'bmi', hue='gender')
ax2.set_title('BMI distribution', fontsize = 14)


ax3 = fig.add_subplot(2,2,3)
sns.kdeplot(data = df , x ='avg_glucose_level')


ax4 = fig.add_subplot(2,2,4)
sns.boxplot(data = df, y = 'age', color = 'grey', saturation=0.1)


plt.show()
```
